# Library

```python
from itertools import *
import seaborn as sns
import matplotlib.pyplot as plt
```

- 목적
  - 회사 지원할 때 자신의 상태가 어느 정도인지 가늠

- 방법
  - 주어진 담당업무, 자격요건, 우대 사항 등 항목 각각의 할 수 있는 정도를 점수화

- level
  - a: 당장 할 수 있는 정도
  - b: 들어봤고 금방 배울 수 있는 정도
  - c: 들어는 봤는데 한 번도 해본 적 없는 정도
  - d: 도움이 많이 필요한 정도

level 마다 점수를 주관적으로 부여한다.  
당장 할 수 있거나, 금방 배울 수 있는 것은 +  
한 번도 해보지 않은 것은 익히는 데에 시간이 걸리므로 0  
도움이 많이 필요한 것은 당장 할 수 있는 것에 완전히 반대되는 상황이므로 -

```python
# level
a = 3
b = 2
c = 0
d = -3

level = [a, b, c, d]
level
```

```python
# 담당업무, 자격요건, 우대사항 등 나눠진 파트마다 가능할 점수 분포를 알아보기 위해
# 중복 순열을 이용해 Num의 수 만큼 각 항목 당 점수를 배분한다.
# 리스트 내부에 만들어진 각 튜플의 합의 히스토그램을 출력한다.
# 이때, 빈도수는 합의 최댓값과 최솟값을 고려한다.
def product_level(Num):
  productList = list(product(level, repeat = Num))
  sumList = []
  for _ in productList:
    sumList.append(sum(_))
  plt.hist(sumList, bins = (max(sumList) - min(sumList) + 1), edgecolor = 'red')
  plt.xlim([min(sumList), max(sumList)])
  plt.title('histogram')
  plt.xlabel('sum')
  plt.ylabel('count')
  plt.show()

  sns.displot(sumList, kde=True, rug=True)
  plt.title('distribution plot')
  plt.xlabel('sum')
  plt.ylabel('count')
  plt.show()

  sns.boxplot(sumList)
  plt.title('boxplot')
  plt.ylabel('sum')
  plt.show()
```

주어진 파트에 항목 갯수에 따라 점수 분포를 확인할 수 있다.  
histogram을 참고하여 점수의 빈도를 확인할 수 있다.  
distribution plot을 참고하여 점수의 분포를 확인할 수 있다.  
boxplot을 참고하여 점수의 분위수가 어디인지 확인할 수 있다.  

```python
product_level(3)
```

주어진 파트에 항목 갯수가 3개이고 본인의 점수가 2점일 경우  
2점의 빈도수는 높은 것으로 보아 나올 수 있는 조합이 많은 것으로 보이며,  
분포나 분위수를 보았을 때 모두 중간보다 살짝 아래에 존재함을 알 수 있다.

```python
product_level(5)
```

주어진 파트에 항목 갯수가 5개이고 본인의 점수가 8점일 경우  
8점의 빈도수는 높은 것으로 보아 나올 수 있는 조합이 많은 것으로 보이며,  
분포나 분위수를 보았을 때 모두 중간보다 위쪽에 존재함을 알 수 있다.

```python
product_level(11)
```
